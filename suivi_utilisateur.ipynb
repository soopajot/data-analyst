```python
import pandas as pd
import numpy as np
df = pd.read_csv('events.csv')
```

```python
from datetime import datetime
df['date'] = df['timestamp'].apply(lambda x: datetime.fromtimestamp(x/1000))
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['dayofweek'] = df['date'].dt.dayofweek
df.head()
```

```python
count = df.groupby('visitorid').agg(['count'])
count = count.sort_values([('timestamp', 'count')], ascending = False)
count.head()
```

```python
stats = pd.DataFrame(count[('timestamp', 'count')].values, columns = ['nb_visites'], index=count[('timestamp', 'count')].index)
stats['nb_transactions'] = pd.DataFrame(count[('transactionid', 'count')].values, index=count[('timestamp', 'count')].index)
stats.head()
```

```python
stats = df.groupby('visitorid').agg({'visitorid':'count'})
stats = stats.rename(columns={"visitorid": "nb_visites"})

view = df.loc[df['event'] == 'view'].groupby('visitorid').agg({'visitorid':'count'})
view = view.rename(columns={"visitorid": "nb_views"})

addtocart = df.loc[df['event'] == 'addtocart'].groupby('visitorid').agg({'visitorid':'count'})
addtocart = addtocart.rename(columns={"visitorid": "nb_addtocarts"})

transaction = df.loc[df['event'] == 'transaction'].groupby('visitorid').agg({'visitorid':'count'})
transaction = transaction.rename(columns={"visitorid": "nb_transactions"})
```

```python
stats = pd.merge(stats, view, left_index=True, right_index=True)
stats = pd.merge(stats, addtocart, left_index=True, right_index=True)
stats = pd.merge(stats, transaction, left_index=True, right_index=True)
stats.head()
```
